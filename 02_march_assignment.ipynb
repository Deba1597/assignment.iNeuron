2:- Scatter plot - A scatter plot shows the relationship between two variables as a collection of points, where each point represents a pair of values. It is commonly used to identify patterns or correlations between variables. Example: plt.scatter(x, y)

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x=3 + np.random.normal(0,2,50)
y=3 + np.random.normal(0,2,len(x))
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y, c=colours , s= sizes , alpha= .67)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('scatter plot')
```

The subplot function in Matplotlib is used to create multiple plots (subplots) in a single figure
it allows Comparing multiple plots,Saving space,Customizing layouts,Sharing axes among the pots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

y2 = np.array([50,20,40,20,60,70])
y3 = np.array([10,20,30,40,50,60])
y4 = np.array([200,350,250,550,450,150])

fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

plt.subplots_adjust(hspace=0.5, wspace=0.5)
plt.show()
```

4:- A bar plot is a type of chart that displays categorical data with rectangular bars, where the height or length of each bar represents the value of the corresponding category. The bars can be horizontal or vertical, and the space between them can be adjusted based on the number of categories and the size of the chart.

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(['Apple','Microsoft','Google','AMD'])
profit = np.array([3000,8000,1000,10000])
plt.bar(company,profit)
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('bar graph')
```

5:- A box plot, also known as a box and whisker plot, is a graphical representation of a dataset that summarizes its distribution and highlights important statistical measures. 
Box plots are useful for comparing distributions between different groups or for visualizing the spread and skewness of a single variable. They are commonly used in fields such as statistics, data analysis, and quality control.

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)

box = [box1,box2] 
fig = plt.figure(figsize =(8, 5))
ax = fig.add_axes([0, 0, 1, 1]) 
bp = ax.boxplot(box)
ax.set_title('Box plot example')
ax.set_xlabel('X-axis label')
ax.set_ylabel('Y-axis label')
plt.show()

```
